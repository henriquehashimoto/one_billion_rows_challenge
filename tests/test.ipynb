```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Dados
data = pd.DataFrame({
    "Ferramenta": ["DuckDB", "PySpark", "FireDucks", "Pandas"],
    "Tempo de Execução (s)": [23.39, 129.84, 141.25, 328.50]
})

# Cores personalizadas
colors = ["#FFA500", "#F4A460", "#B22222", "#B0C4DE"]

# Estilo do Seaborn
sns.set_theme(style="whitegrid")

# Criar gráfico
plt.figure(figsize=(10, 6))
ax = sns.barplot(data=data, x="Ferramenta", y="Tempo de Execução (s)", palette=colors, edgecolor="black")

# Adicionar valores acima das barras
for p in ax.patches:
    ax.annotate(f"{p.get_height():.2f}s", 
                (p.get_x() + p.get_width() / 2, p.get_height() + 10), 
                ha="center", fontsize=12, fontweight="bold")

# Destacar DuckDB
ax.patches[0].set_edgecolor("black")
ax.patches[0].set_linewidth(2)

# Configurações do gráfico
ax.set_title("1 Billion Rows Challenge - Qual mais rápido?", 
             fontsize=16, fontweight="bold", pad=20)
ax.set_ylim(0, 400)

# Ajustar layout e salvar
plt.tight_layout()
plt.savefig("performance_comparison_seaborn.png", dpi=300, bbox_inches="tight")
plt.show()
```
